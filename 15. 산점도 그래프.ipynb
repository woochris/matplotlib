# 15. 산점도 그래프

```python
import matplotlib.pyplot as plt  #라이브러리
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  #windows용
#matplotlib.rcParams['font.family'] = 'AppleGothic'  #mac용
matplotlib.rcParams['font.size'] = 15  #글자 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False  #한글 폰트를 사용할 때 마이너스가 깨지는 현상을 해결
```

```python
import pandas as pd

df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
df['학년'] = [3, 3, 2, 1, 1, 3, 2, 2]
df
```

```python
plt.scatter(df['영어'], df['수학'])
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
```

```python
import numpy as np
sizes = np.random.rand(8) * 1000
sizes
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
```

```python
sizes = df['학년'] * 500

plt.scatter(df['영어'], df['수학'], s=sizes)
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
```

```python
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.4)
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['영어'], df['수학'], s=sizes, c=df['학년'], cmap='viridis', alpha=0.4)
plt.xlabel("영어 점수")
plt.ylabel("수학 점수")
plt.colorbar(ticks=[1, 2, 3], label='학년', shrink=0.5, orientation='horizontal')
```
