# 9. 막대그래프 (심화)

```python
import matplotlib.pyplot as plt  #라이브러리
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  #windows용
#matplotlib.rcParams['font.family'] = 'AppleGothic'  #mac용
matplotlib.rcParams['font.size'] = 15  #글자 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False  #한글 폰트를 사용할 때 마이너스가 깨지는 현상을 해결
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.barh(labels, values)
plt.xlim(175, 195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/')  #////
bar[1].set_hatch('x')  #xxxx
bar[2].set_hatch('..')  #..
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar) :
    plt.text(idx, rect.get_height()+0.5, values[idx], ha='center', color='g')
```
