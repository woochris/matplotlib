# 8. 막대그래프 (기본)

```python
import matplotlib.pyplot as plt  #라이브러리
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  #windows용
#matplotlib.rcParams['font.family'] = 'AppleGothic'  #mac용
matplotlib.rcParams['font.size'] = 15  #글자 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False  #한글 폰트를 사용할 때 마이너스가 깨지는 현상을 해결
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values, color='r')
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b']

plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

plt.bar(labels, values)
plt.ylim(175, 195)  #y축 그래프의 범위제한 175~190
```

```python
plt.bar(labels, values, width=0.5)
```

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45)  #x축의 변수명의 각도를 45도 기울임
plt.yticks(rotation=45)  #xticks랑 똑같음
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]

ticks =['1번학생', '2번학생', '3번학생']

plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=90)
```
