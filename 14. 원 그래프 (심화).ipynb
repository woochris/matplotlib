# 14. 원 그래프 (심화)

```python
import matplotlib.pyplot as plt  #라이브러리
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  #windows용
#matplotlib.rcParams['font.family'] = 'AppleGothic'  #mac용
matplotlib.rcParams['font.size'] = 15  #글자 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False  #한글 폰트를 사용할 때 마이너스가 깨지는 현상을 해결
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ["Python", "Java", "Javasript", "C#", "C/C++", "etc"]
#colors = ["b", "g", "r", "c", "m", "y"]
colors = ['#ffadad', '#ffd6a5', '#fdffb6', '#caffbf', '#9bf6ff', '#a0c4ff']
explode = [0.05]*6

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, explode=explode) 
plt.show()
```

```python
wedgeprops = {"width" : 0.6} #width값의 설정에 따라 값이 0에 가까우면 링모양

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, 
        explode=explode, wedgeprops=wedgeprops) 
plt.show()
```

```python
wedgeprops = {"width" : 0.6, "edgecolor" : "w", "linewidth" : 3} #edgecolor를 흰색으로 해서, explode를 사용한것처럼 눈속임

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops) 
plt.show()
```

```python
def custom_autopct(a) :
    return ('%.1f%%' % a) if a>=10 else ' '

plt.pie(values, labels=labels, autopct=custom_autopct, startangle=90, counterclock=False, colors=colors, wedgeprops=wedgeprops,
       pctdistance=0.7) 
plt.show()
```
