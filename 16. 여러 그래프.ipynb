```python
import matplotlib.pyplot as plt  #라이브러리
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  #windows용
#matplotlib.rcParams['font.family'] = 'AppleGothic'  #mac용
matplotlib.rcParams['font.size'] = 15  #글자 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False  #한글 폰트를 사용할 때 마이너스가 깨지는 현상을 해결
```

```python
import pandas as pd

df = pd.read_excel('../Pandas/score.xlsx')
df
```

```python
fig, axs = plt.subplots(2, 2, figsize=(20, 20)) # 2X2에 해당하는 plots 생성
fig.suptitle("여러 그래프 넣기")

# 첫 번째 그래프
axs[0, 0].bar(df['이름'], df['국어'], label='국어점수')
axs[0, 0].set_title('첫 번째 그래프') #여러 그래프를 그릴 때는 set_title() 
axs[0, 0].legend() 
axs[0, 0].set(xlabel='이름', ylabel='점수') #x축과 y축 라벨 넣을 때도 set()
axs[0, 0].set_facecolor('lightyellow') #전면 색
axs[0, 0].grid(linestyle='--', linewidth=0.5)

# 두 번째 그래프
axs[0, 1].plot(df['이름'], df['수학'], label='수학점수')
axs[0, 1].plot(df['이름'], df['영어'], label='영어점수')
axs[0, 1].legend() 

# 세 번째 그래프
axs[1, 0].barh(df['이름'], df['키'])

# 네 번째 그래프
axs[1, 1].plot(df['이름'], df['사회'], color='g', alpha=0.5)
```
