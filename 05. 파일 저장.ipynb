# 5. 파일저장

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows용
matplotlib.rcParams['font.size'] = 15 #글자 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트를 사용할 때 마이너스가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)

plt.savefig('graph.png', dpi=100)
```

```python
plt.figure(dpi=200)
plt.plot(x, y)
plt.savefig('graph200', dpi=100)
```
