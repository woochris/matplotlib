# 13. 원 그래프(기본)

```python
import matplotlib.pyplot as plt  #라이브러리
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'  #windows용
#matplotlib.rcParams['font.family'] = 'AppleGothic'  #mac용
matplotlib.rcParams['font.size'] = 15  #글자 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False  #한글 폰트를 사용할 때 마이너스가 깨지는 현상을 해결
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ["Python", "Java", "Javasript", "C#", "C/C++", "etc"]

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False) 
# autopct는 퍼센트 나타내는것; 퍼센트비율에 %을 붙이고 싶으면 %%추가
# startangle은 시작 각도
# counterclock = False는 반시계반향
plt.show()
```

```python
values = [30, 25, 20, 13, 10, 2]
labels = ["Python", "Java", "Javasript", "C#", "C/C++", "etc"]
# explode = [0.2, 0.1, 0, 0, 0, 0] # 영역을 분리시켜주는 변수
explode = [0.05]*6

plt.pie(values, labels=labels, explode=explode) #explode은 파이차트의 영역 분리
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode) #explode은 파이차트의 영역 분리
plt.legend(loc=(1.2, 0.3))
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode) #explode은 파이차트의 영역 분리
plt.title("언어")
plt.legend(loc=(1.2, 0.3), title="언어별 선호도")
plt.show()
```
