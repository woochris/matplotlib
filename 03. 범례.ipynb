# 3. 범례 (legend)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows용
matplotlib.rcParams['font.size'] = 15 #글자 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트를 사용할 때 마이너스가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend()
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend(loc='upper right')
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend(loc='lower right')
```

```python
plt.plot(x, y, label='범례')
plt.legend(loc=(0.7, 0.8)) #x축, y축 (0~1사이)
```
