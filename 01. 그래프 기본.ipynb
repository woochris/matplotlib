# Matplotlib
다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

## 1. 그래프 기본

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.show()
```

##  Title 설정

```python
plt.plot(x, y)
plt.title("Line graph")
```

## 한글 폰트 설정

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows용
matplotlib.rcParams['font.size'] = 15 #글자 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트를 사용할 때 마이너스가 깨지는 현상을 해결
```

```python
plt.plot(x, y)
plt.title("꺾은선 그래프")
```
