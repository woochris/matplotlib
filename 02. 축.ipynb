# 2. 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' #windows용
matplotlib.rcParams['font.size'] = 15 #글자 사이즈
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트를 사용할 때 마이너스가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
plt.title("꺾은선 그래프")
```

```python
plt.plot(x, y)
plt.xlabel('x축')
plt.ylabel('y축')
```

```python
plt.plot(x, y)
plt.xlabel('x축', color='red')
plt.ylabel('y축', color='green')
```

```python
plt.plot(x, y)
plt.xlabel('x축', color='red', loc="right") #left, center, right
plt.ylabel('y축', color='green', loc='top') #top, center, bottom
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3])
plt.yticks([3, 6, 9, 12])
plt.show()
```
